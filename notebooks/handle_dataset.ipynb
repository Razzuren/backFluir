# Open Xlsx file

```python
# Import necessary libraries
import pandas as pd
import openpyxl

# Define the path to the Excel file
file_path = '/mnt/c/Users/samue/PycharmProjects/mvp-fluir/dataset/dataset_excel.xlsx'

# Load the Excel file with multiple sheets
excel_data = pd.read_excel(file_path, sheet_name=None)

# Create a dictionary to store DataFrames for each sheet
dataframes = {}

# Get the list of sheet names
sheet_names = list(excel_data.keys())

# Iterate through each sheet and set the first row as the header for all sheets except the last one
for sheet_name in sheet_names[:-1]:
    df = excel_data[sheet_name]
    df.columns = df.iloc[0]  # Set the first row as the header
    df = df[1:]  # Remove the first row from the DataFrame
    dataframes[sheet_name] = df.reset_index(drop=True)  # Reset the index and store the DataFrame

# Add the last sheet without modifying the header
last_sheet_name = sheet_names[-1]
dataframes[last_sheet_name] = excel_data[last_sheet_name]

# Display the dictionary 
print(dataframes)

# Now each sheet is stored as a separate DataFrame in the `dataframes` dictionary
# You can access each DataFrame using the sheet name as the key
```
